### 1. Columns Unnamed 22 and 23, what are they? Can they get dropped?

#### These 2 columns only contain 1 or 2 non-null values and have very ambiguous names. The entries into the database can be submitted by short form with many text fields, which means there is a lot of room for unique, un-categorisable entries.

#### 1.1 Check what the 3 values actually are.
#### 1.2 Decide whether these two columns are worth keeping or dropping

```python
df.loc[~df["Unnamed: 22"].isnull(), "Unnamed: 22"]
```

```python
df.loc[~df["Unnamed: 23"].isnull(), "Unnamed: 23"]
```

#### After some quick research, the value "Teramo" is worth checking, but could simply be the origin of a shark attack victim, as the city is landlocked. If it does not relate to any info in the db, I'll drop the two columns.

```python
df.loc[4415]
```

```python
# Teramo does not seem to be relevant information, hence I will drop the columns

df = df.drop(["Unnamed: 22", "Unnamed: 23"], axis=1)

df
```

### 2. There seems to be a few unpopulated rows, let's double check by displaying max_columns, then give them the boot

```python
pd.set_option('display.max_columns', 30)

df
```

```python
df = df.dropna(how='all')

df
```

#### All completely empty rows have been dropped, but from just taking a look at the above df display, we can see that the column "case number" has some useless 0's and xx's that saved it from being completely empty. I'll take a look at what the comlumn is populated with, and remove the affected rows if all other rows only have NaN.

```python
df.loc[df["Case Number"].isin(["0", "xx"])]
```

```python
# If only the "Case Number" column is populated, but the rest of the row is empty, then it will be deleted.

# df.iloc[:, 1:].isna().sum()

df.isnull().values
```

```python
df.iloc[6308]
```

```python
# Dropping empty_cases rows, because np.all considers 0 to be False, the Case Numbers with 0 will also be removed

empty_cases = df.loc[df["Case Number"].isin(["0", "xx"])]
empty_index = empty_cases.index.to_list()

for row in empty_index:
    if np.all(df.loc[row]) == True:
        df = df.drop(index=row)
    
df
```
